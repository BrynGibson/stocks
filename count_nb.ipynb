```python
def count_post(post):
    mentions = []
    
    
    for ticker, name in zip(tickers_df["symbol"], tickers_df["fixed_name"]):
        ticker = re.escape(ticker)
        if pd.isnull(ticker):
            continue
        if pd.isnull(name):
            exp = r"\$"+ticker+"([,/]|\.|\s)"
        else:
            name = re.escape(name)
            exp = r"(\$"+ticker+"(,|\.|\s)"+"|((,|\s|\$|\.)"+name+r"([,/]|\.|\s))"
            
           
            
        
    
        if re.search(exp, post, flags=re.IGNORECASE | re.MULTILINE):
            mentions.append(ticker)
    
    print("done")
    return str(mentions)
       
    
```
