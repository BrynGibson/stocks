```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

data_path = Path(r"C:\Users\sloth\stonks\data\built")
ticks = ["idex", "amc", "gme", "bb", "uso", "uco", "tsla", "nio"]
```

```python
def test_ticker(ticker, alpha):
    
    print(f"\n\n\nTest results for {ticker}")
    tick_path = data_path / f"{ticker}.csv"
    df = pd.read_csv(tick_path, parse_dates=["Date"], index_col="Date")
    
    df = df.dropna(subset=["Adj Close",])
    
    plt.rcParams["figure.figsize"] = (20,5)



    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_title(f'{ticker} price vs total mentions')
    ax1.set_xlabel('time (d)')
    ax1.set_ylabel('price (usd)', color=color)
    ax1.plot(df["Adj Close"], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('total mentions', color=color)  # we already handled the x-label with ax1
    ax2.plot(df["total_mentions"], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
    
    plt.rcParams["figure.figsize"] = (20,5)



    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_title(f'{ticker} price vs positive mentions')
    ax1.set_xlabel('time (d)')
    ax1.set_ylabel('price (usd)', color=color)
    ax1.plot(df["Adj Close"], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('positive mentions', color=color)  # we already handled the x-label with ax1
    ax2.plot(df["positive_mentions"], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
    
    plt.rcParams["figure.figsize"] = (20,5)



    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_title(f'{ticker} price vs negative mentions')
    ax1.set_xlabel('time (d)')
    ax1.set_ylabel('price (usd)', color=color)
    ax1.plot(df["Adj Close"], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('negative mentions', color=color)  # we already handled the x-label with ax1
    ax2.plot(df["negative_mentions"], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
    
    x = df["total_mentions"]
    y = df["Adj Close"]

    r, p = scipy.stats.pearsonr(x,y)
    
    print(f"\nthe pearson r value is {r}")
    if p < alpha:
        print(f"the pearson p value is {p}, our chosen alpha value is {alpha}, since the p value is less than the alpha value we can reject the null hypothesis with 95% confidence. Therefore it is likely there is correlation between {ticker} and total mentions on reddit.")
    else:
         print(f"the pearson p value is {p}, our chosen alpha value is {alpha}, since the p value is greater than or equal to the alpha value we fail to reject the null hypothesis with 95% confidence. Therefore it is likely there is no correlation between {ticker} and total mentions on reddit.")
        
    r, p = scipy.stats.spearmanr(x,y)
    
    
    print(f"\nthe spearman rho value is {r}")
    if p < alpha:
        print(f"the spearman p value is {p}, our chosen alpha value is {alpha}, since the p value is less than the alpha value we can reject the null hypothesis with 95% confidence. Therefore it is likely there is correlation between {ticker} and total mentions on reddit.")
    else:
         print(f"the spearman p value is {p}, our chosen alpha value is {alpha}, since the p value is greater than or equal to the alpha value we fail to reject the null hypothesis with 95% confidence. Therefore it is likely there is no correlation between {ticker} and total mentions on reddit.")
    
    
    r, p = scipy.stats.kendalltau(x,y)
    print(f"\nthe kendall tau value is {r}")
    
    if p < alpha:
        print(f"the kendall p value is {p}, our chosen alpha value is {alpha}, since the p value is less than the alpha value we can reject the null hypothesis with 95% confidence. Therefore it is likely there is correlation between {ticker} and total mentions on reddit.")
    else:
         print(f"the kendall p value is {p}, our chosen alpha value is {alpha}, since the p value is greater than or equal to the alpha value we fail to reject the null hypothesis with 95% confidence. Therefore it is likely there is no correlation between {ticker} and total mentions on reddit.")
    
```

```python
for ticker in ticks:
    test_ticker(ticker, 0.05)
```
