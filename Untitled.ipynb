```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
data_path = Path(r"C:\Users\sloth\stonks\data\built")
```

```python
ticker = "tsla"
```

```python
tick_path = data_path / f"{ticker}.csv"
```

```python
df = pd.read_csv(tick_path, parse_dates=["Date"], index_col="Date")
```

```python
df
```

```python
plt.rcParams["figure.figsize"] = (20,5)



fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_title(f'{ticker} price vs total mentions')
ax1.set_xlabel('time (d)')
ax1.set_ylabel('price (usd)', color=color)
ax1.plot(df["Adj Close"], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('total mentions', color=color)  # we already handled the x-label with ax1
ax2.plot(df["total_mentions"], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
